```python
#Title: "Wine Quality Prediction Using Vector Machine"

# Ojective : Is to check wine quality if it is acidic or having suger or other parameter concerns

# Data Sources : "https://github.com/YBI-Foundation/Dataset/raw/main/WhiteWineQuality.csv"
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/WhiteWineQuality.csv',sep=';')
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df['quality'].value_counts()
```

```python
df.groupby('quality').mean()
```

```python
y = df['quality']
x = df.drop(['quality'],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
```

```python
x = ms.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
y_pred.shape
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
